Generate a random number between ‘0’ and ‘1’. If that number >= 0.5, then it will count it as heads, otherwise tails. Do this n times using a list comprehension of python. This should happen within the function run_binom.
Repeat this a specified number of times (the input variable trials specify the number of trials). We will perform 1,000 trials.

```python
#n This is 10 times you randomly draw a number
trials_set = 10
#
No_of_trail_sets = 1000
# >=0.5 -- heads, <0.5 tails
threshold = 0.5
```

```python
def exp_binom(No_of_trail_sets,trials_set,threshold):
    result = []
    for i in range(No_of_trail_sets):
        trials_set_result = [np.random.random() for i in range(trials_set)]
        result.append(len([i for i in trials_set_result if i>=threshold]))
    return result
```

```python
head = exp_binom(1000,10,0.5)
```

```python
ax=seaborn.distplot(head,kde=True,color='orange',hist_kws={"linewidth": 22,'alpha':0.77})
ax.set(xlabel='Number of Heads',ylabel='Frequency')
```
