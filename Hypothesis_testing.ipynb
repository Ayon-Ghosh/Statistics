# Difference between Parametric and Non Parametric Test

Parametric tests assume underlying statistical distributions in the data. ... Nonparametric tests do not rely on any distribution. They can thus be applied even if parametric conditions of validity are not met. Parametric tests often have nonparametric equivalents.

If the mean more accurately represents the center of the distribution of your data, and your sample size is large enough, use a parametric test. If the median more accurately represents the center of the distribution of your data, use a nonparametric test even if you have a large sample size

Nonparametric method refers to a type of statistic that does not require that the population being analyzed meet certain assumptions, or parameters. ... Often nonparametric methods will be used when the population data has an unknown distribution, or when the sample size is small

The Chi-square test is a non-parametric statistic, also called a distribution free test. Non-parametric tests should be used when any one of the following conditions pertains to the data: ... The data violate the assumptions of equal variance or homoscedasticity

```python
import scipy.stats
import numpy as np
import matplotlib.pyplot as plt
x= np.linspace(-10, 10, 100)
#degree of freedom df
df = 3.34
#moments : [optional] composed of letters [‘mvsk’]; ‘m’ = mean, ‘v’ = variance, 
#‘s’ = Fisher’s skew and ‘k’ = Fisher’s kurtosis. (default = ‘mv’).
mean, var = scipy.stats.t.stats(df, moments='mv')
print('mean: {:.2f}, variance: {:.2f}'.format(mean, var))
plt.plot(x, scipy.stats.t.pdf(x,df))
plt.show()
#mean: 0.00, variance: 2.49
```

# Example 1

This test will tell us whether the mean of the population and sample are different

t = (sample mean - population mean)/ standard error

where
standard error  = sample standard deviation/ sqr.root of sample size

Null Hypothesis: The mean of the sample and population are same

Alternate Hypothesis: The mean of the sample and population are different


Population of ages

```python
ages = [10,20,35,18,17, 70, 30,76, 12, 18, 23,24,53, 34, 43, 56, 49, 14, 28, 32, 25, 35, 48, 54, 48, 64, 63, 12, 15]
```

```python
import numpy as np
ages_mean = np.mean(ages)
ages_mean
```

Sample of ages

```python
sample_size = 10
age_sample = np.random.choice(ages, sample_size)
age_sample
```

```python
from scipy.stats import ttest_1samp
```

```python
ttest, p_value = ttest_1samp(age_sample, ages_mean)
print(f'ttest:{ttest} p_value:{p_value}')
```

we reject the Alternate Hypothesis and accept the null hypothesis because the p_value is>0.05

# Two Sample T-test with Python 

The independant samples t test or 2-sample t test compares the means of 2 independant groups in order to determine whether there is statistical evidence that the associated population means are significantly different. The independant sample t test is a parametric test.

```python
np.random.seed(10)
college_age = stats.poisson.rvs(loc = 18, mu = 35, size = 1500)
classA_age = stats.poisson.rvs(loc = 18, mu = 30, size = 60)
classB_age = stats.poisson.rvs(loc = 18, mu = 30, size = 60)
```

```python
_,p_value = stats.ttest_ind(a = classA_age, b = classB_age, equal_var = False)
```

# Anova Test(F-Test) One Way F-test(Anova)


The t-test works well when dealing with two groups, but sometimes we want to compare more than two groups at the same time.

For example, if we wanted to test whether petal_width means differs based on some categorical variable like species, we have to compare the means of each level or group the variable

It tells whether two or more groups are similar or not based on their mean similarity and f-score.

Example : there are 3 different category of iris flowers and their petal width and need to check whether all 3 group are similar or not

```python
import seaborn as sns
df=sns.load_dataset('iris')
df.head()
```

```python
df_anova = df[['petal_width','species']]
df_anova.head()
```

```python
df_anova[df_anova['species']=='setosa'].petal_width
```

```python
species_name = df_anova.species.unique()
```

```python
anova_dict = {}
for i in range(len(species_name)):
       anova_dict.update({species_name[i]:df_anova[df_anova['species']==species_name[i]].petal_width})
```

```python
anova_dict
```

```python
F, p = stats.f_oneway(anova_dict['setosa'], anova_dict['versicolor'], anova_dict['virginica'])
print(p)
```

```python
if p<0.05:
    print("reject null hypothesis")
else:
    print("accept null hypothesis")
```

# Chi-Square Test

The test is applied when you have two categorical variables from a single population. It is used to determine whether there is a significant association between the two variables

```python
import scipy.stats as stats
import seaborn as sns
import pandas as pd
import numpy as np
df=sns.load_dataset('tips')
df.head()
```

```python
df_smoker=df.groupby(['sex','smoker'])['smoker'].count()
```

```python
df_smoker.values.reshape(2,2)
```

```python
# or
```

```python
dataset_table=pd.crosstab(df['sex'],df['smoker'])
print(dataset_table)
```

```python
dataset_table.values
```

```python
#Observed Values
Observed_Values = dataset_table.values
print("Observed Values :-\n",Observed_Values)
```

```python
val=stats.chi2_contingency(dataset_table)
```

```python
Expected_Values=val[3]
Expected_Values
```

```python
no_of_rows=len(dataset_table.iloc[0:2,0])
no_of_columns=len(dataset_table.iloc[0,0:2])
ddof=(no_of_rows-1)*(no_of_columns-1)
print("Degree of Freedom:-",ddof)
alpha = 0.05
```

```python
from scipy.stats import chi2
chi_square=sum([(o-e)**2./e for o,e in zip(Observed_Values,Expected_Values)])
chi_square_statistic=chi_square[0]+chi_square[1]
```

```python
zip(Observed_Values,Expected_Values)
```

```python
list(zip(Observed_Values,Expected_Values))
```

```python
print("chi-square statistic:-",chi_square_statistic)
```

ppf is percent point function inverse of cdf - cumulative density function
p = 1 - alpha 
df = degree of freedom

```python
critical_value=chi2.ppf(q=1-alpha,df=ddof)
print('critical_value:',critical_value)
```

```python
#p-value
p_value=1-chi2.cdf(x=chi_square_statistic,df=ddof)
print('p-value:',p_value)
print('Significance level: ',alpha)
print('Degree of Freedom: ',ddof)
print('p-value:',p_value)
```

```python
if chi_square_statistic>=critical_value:
    print("Reject H0,There is a relationship between 2 categorical variables")
else:
    print("Retain H0,There is no relationship between 2 categorical variables")
    
if p_value<=alpha:
    print("Reject H0,There is a relationship between 2 categorical variables")
else:
    print("Retain H0,There is no relationship between 2 categorical variables")
```
